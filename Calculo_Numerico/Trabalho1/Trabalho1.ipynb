```python

def func(x,y):
    f = 3*np.cos(x)*np.cos(y)
    return f
```

```python
# Plotagem da superfície f

x = np.arange(-3.0, 3.1, 0.1)
y = np.arange(-3.0, 3.1, 0.1)
X,Y = np.meshgrid(x,y)
Z = func(X,Y)

fig = plt.figure(figsize=(13,7))
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap='hsv', edgecolor=None)
ax.set(xlabel='$x$', ylabel='$y$', zlabel='$z$', title='f = $3cos(x)cos(y)$')
m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(Z)
plt.colorbar(m)
plt.show()
```

```python
# Curvas de nível de f

fig,ax=plt.subplots(1,1, figsize = (9, 6))
cp = ax.contourf(X, Y, Z, levels=30)
fig.colorbar(m)                                 # Add a colorbar to a plot
ax.set_title('Curvas de Nível de f(x,y)')
#ax.set_xlabel('x (cm)')
#ax.set_ylabel('y (cm)')
plt.show()
```

```python
# Declarando a sequência (mapa logístico):

def seq(x, r):
    return r*x*(1-x)
```

```python
# Definindo a MÉDIA e a VARIÂNCIA

def media(x, n):
    return np.sum(x)/n

def var(x, n, m):

    tmp = 0.0

    for i in range(0, n, 1):
        tmp += (x[i] - m**2)

    return tmp/(n-1)
```

```python
# a = 1.0
m1 = media(x1, n)
sigma1 = var(x1, n, m1)

# a = 2.0
m2 = media(x2, n)
sigma2 = var(x2, n, m2)

# a = 3.8
m3 = media(x3, n)
sigma3 = var(x3, n, m3)

# a = 4.0
m4 = media(x4, n)
sigma4 = var(x4, n, m4)
```

```python
# Printar so resultados de médias e variâncias

print('\nPara a = 1.0, temos a média =', m1, 'e variância =', sigma1)
print('\nPara a = 2.0, temos a média =', m2, 'e variância =', sigma2)
print('\nPara a = 3.8, temos a média =', m3, 'e variância =', sigma3)
print('\nPara a = 4.0, temos a média =', m4, 'e variância =', sigma4)
```

# Exercício D)

No útimo exercício foram feitos diversos passos para analisar o comportamento de uma sequência gerada de um mapa logístico:
\begin{equation}
x_n = a x_{n-1} (1 - x_{n-1})
\end{equation}

Foi considerado o $x_0 = 0.1$, enquanto que o escalar a variou a pontos fixos no intervalo $[0.0, 4.0]$.

Utilizando bibliotecas da *matplotlib*, foi possível plotar quatro gŕaficos para diferentes valores do escalara *'a'*, no que tange a sequência. Nesse contexto, podemos ver que dependendo dos valores do escalar. Por exemplo, para $a = 2$, o comportamento dos resultados tende a um valor constante. Já para $a = 3.8$, tende a diversos resultados.

Ainda com *matplotlib*, geramos os histogramas - com a função *hist()* - para os diferentes escalares nas sequencias. Tais histogramas esclarecem a faixa de resultaddos obtidas para aqueles valores de 'a'.

Agora, utilizamos os recursos do *python* para calcular a média e variância para tal sequência. Seguem, respectivamente as fómulas matemáticas que nos dão tais resultados:

\begin{equation}
  \bar x = \frac{1}{N} \sum_{i=1}^{N}{x_i}
\end{equation}
\begin{equation}
  \sigma = \frac{1}{N - 1} \sum_{i=1}^{N}{(x - \bar x)²}
\end{equation}

Pela célula [16], vê-se que tais resultados foram preparados como funções em *python* e com os recursos da biblioteca *numpy*, de modo qeu tanto a média qunato a variância foram imprimidos em tela para so diferentes valoes do escalar.

Por último, foram usados os recursos já conhecidos de *laçoes em python* e o plot de gráficos pela *matplotlib* para impressão de um Digrama de Bifurcação. Tal diagrama nos retorna os valores da sequência do mapa logistíco em função de todos os valores de escalar dentro de um intevalo limitado.

Perceba, então, que o resultado converge com o esperado à medida que o valor dde 'a' vai se alterando, tal como ficou mais explicitado e exemplificado pelo histogramas.
